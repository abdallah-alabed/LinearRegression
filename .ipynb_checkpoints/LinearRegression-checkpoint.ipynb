# Linear Regression

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Read The Data
> https://www.kaggle.com/jamesbasker/height-weight-single-variable-data-101-series-10

```python
Data = pd.read_csv('Height_Weight_single_variable_data_101_series_1.0.csv')
```

## First 5 rows of data
> Weights in Kg and Heights in CM

> This data was created by Pawan to test linear regression models 
> check the kaggle link : https://www.kaggle.com/testpython/linear-regression

```python
Data.head(5)
```

## Create The Model


```python
X=Data['Weight'].values
X=X.reshape(-1,1)
```

```python
Y=Data['Height'].values
```

## Split the data for training and testing

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=10)
```

```python
X_train=X_train.reshape(-1,1)
X_test=X_test.reshape(-1,1)
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

## Model Scores

```python
train_acc=model.score(X_train,Y_train)
train_acc
```

```python
test_acc=model.score(X_test,Y_test)
test_acc
```

## Testing

#### what is the expected height for 81Kg,55Kg and 110Kg ?

```python
Test_Data=np.array([81,55,110]).reshape(-1,1)
model.predict(Test_Data)
```

## Plotting

#### Plot of the training Data

```python
plt.scatter(X_train,Y_train,color='black')
plt.xlabel('Weight in Kg')
plt.ylabel('Height in CM')
plt.title('Training Data')
```

#### Plot of the testing Data

```python
plt.scatter(X_test,Y_test,color='red')
plt.xlabel('Weight in Kg')
plt.ylabel('Height in CM')
plt.title('Testing Data')
```

#### Plot of Predicted Data with Training Data

```python
X_Extra=np.array([81,55,110,66,102,77,65,88,109,45,39,67]).reshape(-1,1)
Y_Extra=model.predict(X_Extra)
Y_Extra
```

```python
plt.scatter(X_train,Y_train,color='black')
plt.scatter(X_Extra,Y_Extra,color='yellow')
plt.xlabel('Weight in Kg')
plt.ylabel('Height in CM')
plt.title('predicted vs Testing Data')
```
